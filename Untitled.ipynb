```python

# 1 5 2 6 3 4
# 1 2 3 4 5 6

list1 = [1,5,2,6,3,4]
for i in range(len(list1)):
    for j in range(len(list1)-1):
        if list1[j]>list1[j+1]:
            temp = list1[j+1]
            list1[j+1] = list1[j]
            list1[j] = temp
        print(list1)
            
#print(list1)
```

# Selection Sort

```python
# 64 25 12 22 11
# 11 12 22 25 64
list1 = [64,25,12,22,11]
for i in range(len(list1)-1):
    min = list1[i]
    for j in range(i+1,len(list1)):
        if list1[j]<min:
            #print(list1[j])
            min = list1[j]
            min_index = j
            
    temp = list1[i]
    list1[i] = list1[min_index]
    list1[min_index] = temp        
print(list1)
```

```python
import itertools
l2=[]
l3=[]
def hackerCards(arr,k):
    l1=[]
    for i in range(1,k+1):
        if i in arr:
            l1.append(i)

    for L in range(0, len(l1)+1):
          for subset in itertools.combinations(l1, L):
                if sum(list(subset))<=k:
                    l2.append(list(subset))
                    l3.append(len(list(subset)))
    x = max(l3)
    for i in range(len(l2)):
        if x == len(l2[i]):
            return l2[i]
hackerCards(arr,k)
```
